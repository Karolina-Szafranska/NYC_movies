```python
import json
```

```python
def extract_genre_names(genre_string):
    genres_list = json.loads(genre_string.replace("'", '"'))
    genre_names = [genre['name'] for genre in genres_list]
    return ', '.join(genre_names)
```

```python
df_clean_first['genres'] = df_clean_first['genres'].apply(extract_genre_names)
```

```python
df_clean_first.to_csv('cdf_clean_first.csv', index=False)
```

```python
desired_column_order = ['title', 'genres', 'release_year', 'overview','imdb_id', 'budget', 'revenue', 'vote_average', 'vote_count']
```

```python
df_clean_first = df_clean_first[desired_column_order]
```

```python
df_clean_first
```

```python
import re


specific_keywords = ['New York', 'Staten Island', 'Central Park', 'Manhattan', 'Brooklyn', 'Bronx', 'Queens', 'NYC']

# Function to check if any specific keyword is present in the text
def contains_specific_keyword(text):
    if pd.isna(text):  # Check if the text is NaN
        return False
    for keyword in specific_keywords:
        pattern = r'\b{}\b'.format(re.escape(keyword))  # Construct regex pattern with word boundaries
        if re.search(pattern, text, flags=re.IGNORECASE):
            # Ensure that the keyword matches the full name (two words)
            if len(keyword.split()) == 2:  
                return True
    return False

# Apply the function to filter movies
df_clean = df_clean_first[df_clean_first['title'].apply(contains_specific_keyword) |
                              df_clean_first['overview'].apply(contains_specific_keyword)]
```

```python
df_clean
```

```python
df_clean = df_clean.sort_values(by='release_year', ascending=False)
```

```python
df_clean
```

```python
df_clean['release_year'] = pd.to_numeric(df_clean['release_year'], errors='coerce')
```

```python
df_updated = df_clean.dropna(subset=['genres'])

# Split genres into separate rows
df_genres = df_updated.assign(genres=df_updated['genres'].str.split(', ')).explode('genres')

# Group by release decade and genre, and count the number of unique titles for each combination
df_genres['release_decade'] = (df_genres['release_year'] // 10) * 10
genre_title_count = df_genres.groupby(['release_decade', 'genres'])['title'].nunique().unstack(fill_value=0)

# Create a clustered bar chart
plt.figure(figsize=(12, 6))
sns.heatmap(genre_title_count, cmap='Blues', annot=True, fmt='d', linewidths=.5)
plt.title('Number of Titles by Genre and Release Decade')
plt.xlabel('Genre')
plt.ylabel('Release Decade')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by release decade and genre, and count the number of unique titles for each combination
genre_title_count = df_genres.groupby(['genres'])['title'].nunique().sort_values(ascending=False)

# Plot the most popular genres
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_title_count.values, y=genre_title_count.index, palette='viridis')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.title('Most Popular Genres for Movie Titles based in NYC area')
plt.show()
```
